```python
import smtplib, ssl
from email.message import EmailMessage


print ("Selamat datang!")
nama_penerima = []
email_penerima = []
subject = []
pesan = []

def menu():
    print ("\n1.Tulis Email")
    print ("2.Daftar Email")
    print ("3.Tambah Email")
    print ("4.Keluar")

def tambah():
    print ("Ingin Menambahkan Email Kembali?")
    print ("1.Ya")
    print ("2.Tidak")
def tambah_email():
    nama = str(input("Masukkan Nama Penerima: "))
    email = str(input("Masukkan Email Penerima : "))
    nama_penerima.append(nama)
    email_penerima.append(email)
    print("Email berhasil ditambahkan\n")
def create_contact():
    if not nama_penerima:
        print ("Daftar Kontak KOSONG!\nSilahkan Menambahkan Email")
    else:
        f = open("Receiver_List.txt", "w")
        for x in range(len(nama_penerima)):
            f.write("{}. Nama: {} \n   Email: {} \n".format(x+1,nama_penerima[x],email_penerima[x])) 
        f.close()

def pesan_email():
    subject_email = str(input("Masukkan Subject Email: "))
    pesan_email = str(input("Masukkan Pesan Email: "))
    subject.append(subject_email)
    pesan.append(pesan_email)
            
def kirim_email():
    port = 587  # For starttls
    smtp_server = "smtp.gmail.com"
    sender_email = "FinalProjectBasicPython.2021@gmail.com" #dummy email
    receiver_email = "adri.prasetyo.1992@gmail.com"
    password = "Final_Project_2021"
    subjects = (subject[0])
    pesans = (pesan[0])
    
    msg = EmailMessage()
    msg.set_content(pesans)

    msg['Subject'] = subjects
    msg['From'] = sender_email
    msg['To'] = email_penerima

    context = ssl.create_default_context()
    with smtplib.SMTP(smtp_server, port) as server:
        server.ehlo() 
        server.starttls(context=context)
        server.login(sender_email, password)
        server.send_message(msg)
        

while True:
    menu()
    pilih_menu = int(input("Pilih Menu :\n(masukkan angka saja) "))
    if pilih_menu == 1:
        if not nama_penerima:
            print ("Daftar Email KOSONG!\nSilahkan Menambahkan Email")
        else:
            pesan_email()
            #kirim_email()
            #create_contact()
            kirim = int(input("Kirim email?\n1.Ya\n2.Tidak\n"))
            if kirim == 1:
                kirim_email()
                create_contact()
                print("email berhasil dikirim")
            else:
                create_contact()
                print("")
    elif pilih_menu == 2:
        if not nama_penerima:
            print ("Daftar Email KOSONG!\nSilahkan Menambahkan Email")
        else:
            for x in range(len(nama_penerima)):
                print("{}. Nama: {} \n   Email: {} \n".format(x+1,nama_penerima[x],email_penerima[x])) 
    elif pilih_menu == 3:
        while True:
            tambah_email()
            tambah()
            lanjut = int(input("Pilih 1/2 \n"))
            if lanjut == 1:
                print("")
            else:
                create_contact()
                break  
    elif pilih_menu == 4:
        print("Terima Kasih telah mencoba program kami :D")
        break             
    else: 
        print ("Menu tidak tersedia")

        
#https://docs.python.org/3/library/email.message.html
#https://docs.python.org/3/library/email.message.html
#https://www.freecodecamp.org/news/send-emails-using-code-4fcea9df63f/
#https://stackoverflow.com/questions/7232088/python-subject-not-shown-when-sending-email-using-smtplib-module
```
