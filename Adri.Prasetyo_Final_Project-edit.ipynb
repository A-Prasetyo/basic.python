```python
import smtplib, ssl
from email.message import EmailMessage


print ("Selamat datang!")
nama_penerima = []
email_penerima = []
subject = []
pesan = []

def menu(): #Function untuk menampilkan menu program
    print ("\n1.Tulis Email")
    print ("2.Daftar Email")
    print ("3.Tambah Email")
    print ("4.Keluar")

def tambah_email(): #Function untuk menambahkan penerima email
    nama = str(input("Masukkan Nama Penerima: "))
    email = str(input("Masukkan Email Penerima : "))
    nama_penerima.append(nama)
    email_penerima.append(email)
    print("Email berhasil ditambahkan\n")    
    
def tambah(): #Function untuk kembali menambahkan penerima email
    print ("Ingin Menambahkan Email Kembali?")
    print ("1.Ya")
    print ("2.Tidak")
      
def create_contact(): #Function untuk membuat file.txt dari email yang sudah ditambahkan
    if not nama_penerima:
        print ("Daftar Kontak KOSONG!\nSilahkan Menambahkan Email")
    else:
        f = open("Receiver_List.txt", "w")
        for x in range(len(nama_penerima)):
            f.write("{}. Nama: {} \n   Email: {} \n".format(x+1,nama_penerima[x],email_penerima[x])) 
        f.close()

def pesan_email(): #Function untuk memasukkan Subjek dan Pesan Email
    subject_email = str(input("Masukkan Subject Email: ")) #sebagai Subjek
    pesan_email = str(input("Masukkan Pesan Email: ")) #sebagai isi pesan (body email)
    subject.append(subject_email) #menambahkan string subjek ke list subject *sebenernya bisa pakai return tapi saya gagal terus :D
    pesan.append(pesan_email) #menambahkan string body email ke list body email *sebenernya bisa pakai return tapi saya gagal terus :D
            
def kirim_email(): #Function untuk execute/kirim email 
    #Function menggunakan SMTP yang sudah disediakan python
    #Disini saya menggunakan Start TLS (bisa Pakai SSL)
    port = 587  # Nomor PORT untuk start TLS
    smtp_server = "smtp.gmail.com" #Saya menggunakan SMTP domain gmail.com
    sender_email = "FinalProjectBasicPython.2021@gmail.com" #dummy email untuk mengirim Email
    password = "Final_Project_2021" #Password Dummy Email 
    subjects = (subject[0]) #Untuk memanggil String Subjek yang telah ada di List Subjek
    pesans = (pesan[0]) #Untuk memanggil String Body Email yang telah ada di List Pesan
    
    msg = EmailMessage() #Format Email Message, untuk memudahkan dalam memisahkan Subjek, Body Email, Penerima, dan Pengirim
    msg.set_content(pesans) #String Body Email dimasukkan kedalam sini)
    msg['Subject'] = subjects #String Subjek Email
    msg['From'] = sender_email #Pengirim Email
    msg['To'] = email_penerima #Penerima Email
    #return msg untuk mengecek format EmailMessage
    
    #untuk connect dari program python ke server
    context = ssl.create_default_context()
    with smtplib.SMTP(smtp_server, port) as server:
        server.ehlo() 
        server.starttls(context=context)
        server.login(sender_email, password) #untuk memasukkan email pengirim dan password 
        server.send_message(msg) #untuk mengirimkan pesan
        

        
#Looping Program Pengiriman Email
while True:
    menu()
    pilih_menu = int(input("Pilih Menu :\n(masukkan angka saja) "))
    if pilih_menu == 1:
        if not nama_penerima:
            print ("Daftar Email KOSONG!\nSilahkan Menambahkan Email")
        else:
            pesan_email()
            kirim = int(input("Kirim email?\n1.Ya\n2.Tidak\n"))
            if kirim == 1:
                kirim_email()
                create_contact()
                print("\nEmail berhasil dikirim!\n")
            else:
                create_contact()
                print("")
    elif pilih_menu == 2:
        if not nama_penerima:
            print ("Daftar Email KOSONG!\nSilahkan Menambahkan Email")
        else:
            for x in range(len(nama_penerima)):
                print("{}. Nama: {} \n   Email: {} \n".format(x+1,nama_penerima[x],email_penerima[x])) 
    elif pilih_menu == 3:
        while True:
            tambah_email()
            tambah()
            lanjut = int(input("Pilih 1/2 \n"))
            if lanjut == 1:
                print("")
            else:
                create_contact()
                break  
    elif pilih_menu == 4:
        print("Terima Kasih telah mencoba program kami :D")
        break             
    else: 
        print ("Menu tidak tersedia")

        
#https://docs.python.org/3/library/smtplib.html
#https://docs.python.org/3/library/email.message.html
#https://www.freecodecamp.org/news/send-emails-using-code-4fcea9df63f/
#https://stackoverflow.com/questions/7232088/python-subject-not-shown-when-sending-email-using-smtplib-module
```
